## Import the required modules

```python
import matplotlib.pyplot as plt
import polars as pl
from polars import Expr, LazyFrame, DataFrame
import numpy as np
from pathlib import Path
from typing import Any, Dict, Callable
from QMAP_tools.qmap_inspector import QMAPInspector
from tqdm import tqdm
```

```python
qm = QMAPInspector()
# Load the helper class which by default has the standard relative paths of the database files so should work out of the box
# in most cases, however if the data is stored else where just use the init kwargs to set the correct path
```

```python
qm.full_db.filter(pl.col("i") == 1).collect()
# The full database contains all the data needed to make the agg_db and some. It is very important that you do not try load
# the full database into memory so when previewing a standard way is to just filter by pl.col("i") == 1 which restricts the
# database to 1 / 8192 of its full size and therefore can be fully loaded into memory. We are still able to do operations
# on this dataset as you will see later leveraging the LazyFrame.
```

### Investigate detailed_survey data

```python
ds_columns_of_interest: list[str] = [
    # Raw reflectance
    # "uint8_reflectance",
    "32bit_reflectance",

    # TIR spectral
    # "TIR detailed_survey Slope1000",
    "TIR detailed_survey BD350",
    "TIR detailed_survey BD440",
    "TIR detailed_survey Ratio1000",

    # "TIR detailed_survey sigma Slope1000",
    # "TIR detailed_survey sigma BD350",
    # "TIR detailed_survey sigma BD440",

    # VNIR spectral
    "VNIR detailed_survey BandArea3200to3600nm",
    "VNIR detailed_survey OH2700nm",
    "VNIR detailed_survey Pyroxene920nm",
    "VNIR detailed_survey Slope1polyfit",
    "VNIR detailed_survey Slope2polyfit",

    # "VNIR detailed_survey sigma BandArea3200to3600nm",
    # "VNIR detailed_survey sigma OH2700nm",
    # "VNIR detailed_survey sigma Pyroxene920nm",
    # "VNIR detailed_survey sigma Slope1polyfit",
    # "VNIR detailed_survey sigma Slope2polyfit",
]
```

```python
from itertools import combinations
from matplotlib import colors

for recon in ["recona", "reconb", "reconc"]:
    recon_columns_of_interest: list[str] = [
        column.replace("detailed_survey", recon)
        for column in ds_columns_of_interest
    ]

    ## Look at 1D histograms
    hist_1D_export_folder = Path(f".plots/{recon}_1d_hists")
    hist_1D_export_folder.mkdir(exist_ok=True, parents=True)

    for column in tqdm(recon_columns_of_interest, desc=f"Rendering 1D hists for mission phase {recon}"):
        try:
            
            export_path = hist_1D_export_folder / f"{column}.png"

            if export_path.exists():
                continue

            # Find the range from the database
            range = qm.get_column_extrema(column, qm.full_db.filter(pl.col(column) != -9999.0))

            # Get the counts and bins for a linear binned hist (custom LazyFrame can be used if entered in the kwarg)
            counts, bins = qm.get_column_hist(
                column = column,
                bin_range = range,
                bin_number = 100,
                custom_lf = qm.full_db.filter(pl.col(column) != -9999.0)
            )

            # Plot the histogram data
            plt.bar(
                bins[:-1],
                counts,
                width=np.diff(bins),
                align="edge",
            )

            plt.yscale("log")
            plt.xlabel(column)
            plt.ylabel("Count")
            plt.tight_layout()
            plt.savefig(export_path)
            plt.close()
        except:
            print(f"Rendering 1D hists for {column} failed")

    hist_2D_export_folder = Path(f".plots/{recon}_2d_hists")
    hist_2D_export_folder.mkdir(exist_ok=True, parents=True)
    n_jobs: int = len(ds_columns_of_interest) * (len(ds_columns_of_interest) - 1) // 2

    for x_item, y_item in tqdm(
        combinations(recon_columns_of_interest, 2),
        desc=f"Rendering 2D hists for mission phase {recon}",
        total=n_jobs):
        
        try:
            export_path = hist_2D_export_folder / f"{x_item} vs {y_item}.png"
            if export_path.exists():
                continue

            filtered_lf = qm.full_db.filter(pl.col(x_item) != -9999.0, pl.col(y_item) != -9999.0)

            # Get the hist data
            counts, x_edges, y_edges = qm.get_column_hist2d(
                columns=(x_item, y_item),
                bin_ranges=(
                    qm.get_column_extrema(x_item, filtered_lf), 
                    qm.get_column_extrema(y_item, filtered_lf)
                    ),
                bin_numbers=(100, 100),
                custom_lf=filtered_lf
            )

            plt.figure(figsize=(8, 6))

            plt.pcolormesh(
                x_edges,
                y_edges,
                counts.T,        # transpose because pcolormesh expects (y, x)
                shading="auto",
                cmap="RdYlGn_r",
                norm = colors.LogNorm()
            )

            plt.xlabel(x_item)
            plt.ylabel(y_item)
            plt.colorbar(label = r"Area ($m^2$)")
            plt.tight_layout()
            plt.savefig(export_path)
            plt.close()

        except:
            print(f"Failed for {x_item} vs {y_item}")
            continue
```
